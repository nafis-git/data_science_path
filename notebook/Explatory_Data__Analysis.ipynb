To **fill the nan** values there are **two bellow methods**
- using transform


```python
import pandas as pd

# Load the data
planes = pd.read_csv('https://raw.githubusercontent.com/nafis-git/data_science_path/main/data/planes.csv')

# Find threshold for dropping columns
threshold = len(planes) * 0.05

# Find columns with number of NaNs less than threshold
drop_cols = planes.columns[planes.isna().sum() < threshold]

# Drop rows with NaNs in selected columns
planes.dropna(subset=drop_cols, inplace=True)

# Grouped median of the airline and total stops; The transform function applies the grouped median to each row in the original DataFrame, resulting in a Series of median values with the same length as the original DataFrame.
price_med = planes.groupby(['Airline', 'Total_Stops'])['Price'].transform('median')

# Fill missing values in 'Price' column with grouped median
planes['Price'].fillna(price_med, inplace=True)
```

```python
planes['Additional_Info'] = planes['Additional_Info'].fillna('No info')
```

capturing the correct values by using the ^ symbol to indicate the start of the string

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Load the data
planes = pd.read_csv('https://raw.githubusercontent.com/nafis-git/data_science_path/main/data/planes.csv')

# Fill missing values in the 'Duration' column
planes.dropna(subset = ['Duration'] , inplace=True)

# Create a list of categories
flight_categories = ["Short-haul", "Medium", "Long-haul"]

# Create short_flights
short_flights = "^0h|^1h|^2h|^3h|^4h"

# Create medium_flights
medium_flights = "^5h|^6h|^7h|^8h|^9h"

# Create long_flights
long_flights = "10h|11h|12h|13h|14h|15h|16h"


# Create conditions for values in flight_categories to be created
conditions = [
    (planes["Duration"].str.contains(short_flights)),
    (planes["Duration"].str.contains(medium_flights)),
    (planes["Duration"].str.contains(long_flights)
     )
]

# Apply the conditions list to the flight_categories
planes["Duration_Category"] = np.select(conditions, flight_categories, default="Extreme duration")

# Plot the counts of each category
sns.countplot(data=planes, x="Duration_Category", hue = 'Duration_Category')
plt.show()
```

**Outliers**
IQR= difference between the 75th and 25th percentiles.  
upper_outliers = 75th percentile + 1,5\*IQR.   
Lower outliers =  25th percentile - 1.5\*IQR.  

```python
salary = pd.read_csv('https://raw.githubusercontent.com/nafis-git/data_science_path/main/data/ds_salaries_clean.csv')
salary.head()
```

**calculate the outliers**

```python

# 75th percentile
seventy_fifth = salary['Salary_USD'].quantile(0.75)
# 25th percentile
twenty_fifth = salary['Salary_USD'].quantile(0.25)
iqr = seventy_fifth - twenty_fifth
iqr
```

```python
upper_threshold = seventy_fifth +(1.5*iqr)
lower_threshold = twenty_fifth -(1.5*iqr)
```

```python
lower_threshold #lower limit is actually below zero, which isn't possible given we are working with salaries!
```

```python
# just for the sake of demostration
salary[((salary['Salary_USD' ]> upper_threshold) | (salary['Salary_USD'] < lower_threshold))]\
 [['Experience', 'Employee_Location','Salary_USD']]
```

**Why look for outliers?**
- Outliers are extreme values
> may not accurately represent our data
- Can change the mean and standard deviation
- Statistical tests and machine learning models need normally distributed data


```python
#
salary.describe()
```

```python
# removing outliers
salary[((salary['Salary_USD' ]< upper_threshold) & (salary['Salary_USD'] > lower_threshold))].describe()
```

## **method for looking at class frequency is cross-tabulation**
> enables us to examine the frequency of combinations of classes.

```python
import pandas as pd

# Load the data
planes = pd.read_csv('https://raw.githubusercontent.com/nafis-git/data_science_path/main/data/clean_plane.csv', index_col = 0,
                     parse_dates = ['Date_of_Journey', 'Dep_Time'	, 'Arrival_Time'])
planes.head(2)
```

```python
pd.crosstab(planes['Source'], planes['Destination'])
```

```python
# find the median price for all the flight routes in india
planes.groupby(['Source','Destination'])['Source','Destination'].value_counts().unstack()
```

```python
# find the median price for all the flight routes in india
price_df = planes.groupby(['Source','Destination'])['Price'].median().reset_index()
price_df
```

```python
# or we can do it throughthis method
pd.crosstab(planes['Source'], planes['Destination'], values = planes['Price'], aggfunc = 'median')
```

```python
planes.dtypes
```

```python
# change duration to number of minutes
# Convert duration to number of minutes
planes['Duration'] = planes['Duration'].str.replace('h', '*60+')
planes['Duration'] = planes['Duration'].str.replace('m', '')
# Add a default value of 0 after replacing 'h' with '*60+' to ensure completeness
planes['Duration'] = planes['Duration'].str.replace(r'(\d+)\+\s*$', r'\1+0')  #the pattern (\d+)\+\s*$ matches a sequence of one or more digits followed by a plus sign (+), optional whitespace characters, and the end of the string.
#r'\1+0': This is a backreference to the first captured group ( ) in the pattern,  which is the sequence of digits. It represents the value captured by (\d+) in the pattern.
```

```python
# Convert duration to number of minutes
planes['Duration'] = planes['Duration'].apply(lambda x:eval(x))
```

```python
planes.dtypes
```

```python
# change Total_Stops to numerical values
planes['Total_Stops'].value_counts()
```

```python
# we should change non-stop to 0 and then chnage them to integer
planes['Total_Stops']= planes['Total_Stops'].str.replace(' stops','')  # lights with two, three, or four stops are ready to convert
planes['Total_Stops']= planes['Total_Stops'].str.replace(' stop','')  # lights with two, three, or four stops are ready to convert
planes['Total_Stops'] = planes['Total_Stops'].str.replace('non-stop','0')
```

```python
# convert to integer
planes['Total_Stops'] = planes['Total_Stops'].astype(int)
```
